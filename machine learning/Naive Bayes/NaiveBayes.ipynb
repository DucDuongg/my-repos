```octave
#function, output theta, mu, sigma

# input : 
# X - matrix of size N x D 
# y - label vector of size N x 1

# output: 
# theta - a vector of size K x 1 (prior distribution)
# mu - a matrix K x D
# sigma - a matrix K x D

function [theta, mu, sigma] = train(X, y)
    
    K = length( unique(y) );
    [N, D] = size(X);

    theta = zeros(K, 1);
    mu = zeros(K,D);
    sigma = zeros(K,D);
    
    for k = 1:K
        theta(k) = sum(y == k);
        X_k = X( y == k, : );
        mu(k, :) = mean(X_k);
        sigma(k, :) = std(X_k);
    end
    
    theta = theta / N; # prior probability
    
end
```

```octave
[theta, sigma, mu] = train(X, y)
```
